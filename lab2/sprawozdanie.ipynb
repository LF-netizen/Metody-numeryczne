***Zadanie 1.***  
Zdefiniuj w notatniku funkcję $f(x)= x^{3} - 3x$ i sporządź jej wykres dla argumentów z przedziału:  
1. $x\in<-1,1>$  
2. $x\in<-5,5>$  
3. $x\in<0,5>$
  
***Wskazówki***
Jako argumentu funkcji można użyć numpy array.  

Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/). Działanie tego pakietu prezentuje przykład:

```python
x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
plt.plot(x, y, label ='sin(x)')
plt.show()
```

Wspierając się dokumentacją [Matplotlib](https://matplotlib.org/) dodaj do wykresu etykiety osi, tytuł, grida i legendę.

```python
for left, right in [(-1, 1), (-5, 5), (0, 5)]:
    x = np.linspace(left, right)
    y = x * x * x - 3 * x
    plt.plot(x, y)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.grid(True)
    plt.title('Wykres funkcji f w zakresie ({}, {})'.format(left,right))
    plt.legend(['f(x) = x^3 - 3x'])
    plt.show()
```

Wyraźnie widać, że biblioteka $Matplotlib$ jest wzorowana na tworzeniu wykresów w Matlabie. Fakt programowania w Pythonie pozwala znacznie uprościć wiele powtarzalnych operacji w sposób przejrzysty i intuicyjny. Dokumentacja $Matplotlib$ jest bogata i czytelna, nie brakuje w niej też przykładów użycia danych funkcji czy gotowych, przykładowych wykresów wraz z generującym je kodem.

***Zadanie 2.***  
Dla funkcji z zadania pierwszego używając wektora $x \in <-10,10>$ wyrysuj wykres dla argumentów:
1. $x\in<-1,1>$  
2. $x\in<-10,-1>$  
3. $x\in<1,10>$

Używając do tego funkcji: [ylim](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.ylim.html), [xlim](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xlim.html), [axis](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.axis.html)

```python
x = np.arange(-10, 10, 0.1)
y = x * x * x - 3 * x

plt.xlim(-1, 1)
plt.ylim(-2, 2)
plt.plot(x, y)
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.legend(['f(x) = x^3 - 3x'])
plt.show()

plt.xlim(-10, -1)
plt.ylim(-1000, 0)
plt.plot(x, y)
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.legend(['f(x) = x^3 - 3x'])
plt.show()
  
plt.xlim(1, 10)
plt.ylim(0, 1000)
plt.plot(x, y)
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.legend(['f(x) = x^3 - 3x'])
plt.show()  
```

Funkcje $xlim$ oraz $ylim$ pozwalają na zawężenie bądź rozszerzenie zakresu liczbowego osi na wykresach. Czasem jednak zdarza się, że po zmianie zakresu osi $x$ zakres osi $y$ dobierany automatycznie jest niepoprawny, przez co nieefektywnie wykorzystuje dostępną przestrzeń wykresu, a nawet sprawia, że jest on nieczytelny. W takich sytuacjach konieczne jest ręczne ustawienie zakresu obu osi.

***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

```python
m = 2.5
v = 60 / 3.6
Q = m * v * v / 2
print('Dla ciała o masie 2500g poruszającego się z prędkością 60 km/h po przeliczeniu jednostek i podstawieniu liczb do wzoru otrzymujemy\n Q = {:.2f} J\n'.format(Q))
print('Uwzględniając, iż 1 kcal = 4841 J: \n Q = {:.2f} kcal\n'.format(4841 * Q))

m = 3
v = np.linspace(200, 0)

print('Dla ciała o masie 3000g zależnośc ciepła, które wydzieli się podczas hamowania od prędkości tego ciała jest zobrazowana na poniższych wykresach')

Q = m * v * v / 2 

fig, ax = plt.subplots()
fig.suptitle('Wykres w skali liniowej')
ax.plot(v, Q)
ax.set_xlim(200, 0)
ax.set_ylim(0, 65000)
ax.set_xlabel('v')
ax.set_ylabel('Q')
ax.grid('on')

fig, ax = plt.subplots()
fig.suptitle('Wykres w skali logarytmicznej')
ax.semilogy(v, Q)
ax.set_xlim(200, 0)
ax.set_xlabel('v')
ax.set_ylabel('Q')
ax.grid(which='minor', alpha=0.4)
ax.grid(which='major', alpha=1)
```

$Matplotlib$ pozwala na tworzenie wykresów w skalach logarytmicznych, które $\textit{notabene}$ dobrze współgrają z pewnymi zestawami danych. Dodatkowe linie siatki na wykresie logarytmicznym sprawiają, że ławiej odczytać dane. Należy jednak pamiętać, że skala logarytmiczna jest zazwyczaj mniej intuicyjna, trudniejsza w odbiorze dla laików, przez co osoba niedoświadczona może błędnie zinterpretować dane przedstawione na wykresie. 

***Zadanie 4.***  
Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
```

Pierwsza ma rysować się w kolorze niebieski i grubość lini ma wynosić 4, druga natomiast w kolorze czerwonym o grubości lini 2. Domyślnie ma być dodawana legenda. Dodatkowo użytkownik ma mieć możliwość ustawienia parametrów:

Nie zmienia to jednak faktu, że rozwiązywanie równań graficznie pozostaje co najwyżej ciekawostką, metodą niedokładną, wysoce niepraktyczną. ***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
from main import parallel_plot

phi = np.arange(-50, 30, 0.1)
r = 3 * np.exp(0.07 * phi)
x = r * np.cos(phi)
y = r * np.sin(phi)

x1 = x[:len(x) // 5]
y1 = y[:len(y) // 5]

parallel_plot(x, y, x1, y1, 'x', 'y', 'x', 'y', 'Spirala logarytmiczna', '|')
```
